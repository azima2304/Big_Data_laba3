## Задание 2

Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей и заголовка фигуры.

   Была взята функция вида:

   $$
   y = x^2 - 2x - 8
   $$

### Описание

- **Квадратичная функция**: Это функция вида $y = x^2 - 2x - 8$, где \( a, b, c \) — коэффициенты.
- **Масштаб**: Выбор масштаба осей может быть произвольным, чтобы лучше отразить характер графика.
- **Подписи осей**: Необходимо указать метки для осей X и Y для лучшего понимания графика.
- **Заголовок фигуры**: Заголовок должен быть информативным и отражать суть графика.


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Comic Sans MS'
mpl.rcParams['font.size'] = 12
mpl.rcParams['font.weight'] = 'bold'


fig, ax = plt.subplots(figsize=(10, 6))
ax.grid(color='#EFB509', alpha=0.5)

x = np.linspace(-5, 7, 100)
y = np.power(x, 2) - 2 * x - 8

ax.plot(x, y, color='#002C54', linestyle='-',
        linewidth=2, label='y = x² - 2x - 8')

plt.title('Quadratic Function: y = x² - 2x - 8', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)


ax.grid(True)
ax.legend()
plt.savefig('quadratic_function.png', format='png')
plt.show()
```

## Задание 3

Построить график произвольной гармонической функции, такой как

$$
y = A \sin(Bx + C) + D,
$$

с произвольным масштабом и настроенными подписями осей и заголовком. 

В этом примере мы возьмём гармоническую функцию: 

$$
y_1(x) = \sin(x) + \frac{1}{2} \sin(2x) + \frac{1}{4} \sin(3x)
$$
$$
y_2(x) = \cos(x) + \frac{1}{2} \cos(2x) + \frac{1}{4} \cos(3x)
$$

**Шаги решения:**
1. Определить диапазон значений \( x \) от 0 до \( 2\pi \).
2. Построить графики обеих функций на одном поле.
3. Настроить оси и подписи, используя LaTeX для отображения математических символов.
4. Добавить сетку и легенду для удобного восприятия графиков.

```python
import matplotlib.pyplot as plt
import numpy as np

# Настройка осей
plt.axis([0, 2 * np.pi, -2, 2])
mpl.rcParams['font.family'] = 'Comic Sans MS'
mpl.rcParams['font.size'] = 12
mpl.rcParams['font.weight'] = 'bold'

# Метки на оси X
plt.xticks(np.linspace(0, 2 * np.pi, 9),
           ('0', r'$\frac{1}{4}\pi$', r'$\frac{1}{2}\pi$', r'$\frac{3}{4}\pi$', r'$\pi$',
            r'$\frac{5}{4}\pi$', r'$\frac{3}{2}\pi$', r'$\frac{7}{4}\pi$', r'$2\pi$'),
           fontsize=20)

# Подписи осей
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Заголовок
plt.title(r'Harmonic Functions', fontsize=20)

# Генерация данных
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x) + 0.5 * np.sin(2 * x) + 0.25 * np.sin(3 * x)
y_cos = np.cos(x) + 0.5 * np.cos(2 * x) + 0.25 * np.cos(3 * x)

# Построение графиков
plt.plot(x, y_sin, linewidth=2, color='b', dashes=[
         8, 4], label=r'$\sin(x) + \frac{1}{2}\sin(2x) + \frac{1}{4}\sin(3x)$')
plt.plot(x, y_cos, linewidth=2, color='r', dashes=[
         8, 4, 2, 4], label=r'$\cos(x) + \frac{1}{2}\cos(2x) + \frac{1}{4}\cos(3x)$')

# Легенда
plt.legend(fontsize=15)

# Сетка
plt.grid(True)

# Отображение графиков
plt.show()
```

Построить два графика: квадратичную функцию и гармонические функции синуса и косинуса на одной координатной плоскости.  
Добавить к каждому графику подписи и легенду, используя библиотеку Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib as mpl

# Настройки шрифтов
mpl.rcParams['font.family'] = 'Comic Sans MS'
mpl.rcParams['font.size'] = 12
mpl.rcParams['font.weight'] = 'bold'

# Создание фигуры
fig, ax1 = plt.subplots(figsize=(10, 6))

# Генерация данных для квадратичной функции
x1 = np.linspace(-5, 7, 100)
y1 = np.power(x1, 2) - 2 * x1 - 8

# Построение графика квадратичной функции
ax1.plot(x1, y1, color='#002C54', linestyle='-',
         linewidth=2, label='y = x² - 2x - 8')
ax1.set_title('Quadratic and Harmonic Functions', fontsize=16)
ax1.set_xlabel('X-axis', fontsize=14)
ax1.set_ylabel('Quadratic Function', fontsize=14)
ax1.grid(True)
ax1.grid(color='#EFB509', alpha=0.5)
ax1.legend(loc='upper left')

# Создание второй оси Y
ax2 = ax1.twinx()

# Генерация данных для гармонических функций
x2 = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x2) + 0.5 * np.sin(2 * x2) + 0.25 * np.sin(3 * x2)
y_cos = np.cos(x2) + 0.5 * np.cos(2 * x2) + 0.25 * np.cos(3 * x2)

# Построение графиков гармонических функций
ax2.plot(x2, y_sin, color='#002C54', linestyle=':', linewidth=2,
         label=r'$\sin(x) + \frac{1}{2}\sin(2x) + \frac{1}{4}\sin(3x)$')
ax2.plot(x2, y_cos, color='r', linestyle='--', linewidth=2,
         label=r'$\cos(x) + \frac{1}{2}\cos(2x) + \frac{1}{4}\cos(3x)$')

# Настройка оси Y для гармонических функций
ax2.set_ylabel('Harmonic Functions', fontsize=14)
ax2.legend(loc='upper right')

# Сохранение и отображение графиков
plt.tight_layout()
plt.savefig('combined_graphs_individual_scales.png', format='png')
plt.show()
```

## Задание 6

Построить график функции, заданной параметрически:

$$
\begin{align*}
x(t) &= \cos(3t), \\
y(t) &= \sin(2t), \\
t &\in [0, 4\pi].
\end{align*}
$$

### Описание

- **Параметрическая функция**: Данная функция описывается двумя уравнениями, где \( x \) и \( y \) зависят от параметра \( t \).
- **Диапазон параметра \( t \)**: Параметр \( t \) изменяется от \( 0 \) до \( 4\pi \), что позволяет увидеть полный цикл поведения функции.
- **График**: Визуализируйте зависимость \( y \) от \( x \) для понимания формы и поведения параметрической функции.
- **Настройка графика**: Важно добавить подписи осей и заголовок для лучшего восприятия информации.


```python
import matplotlib.pyplot as plt
import numpy as np

# Определяем диапазон t
t = np.linspace(0, 4 * np.pi, 1000)  # 1000 точек от 0 до 4π

# Задаем параметры x и y
x = np.cos(3 * t)  # x(t) = cos(3t)
y = np.sin(2 * t)  # y(t) = sin(2t)

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='#002C54', linewidth=2,
         label='Параметрическая кривая: x(t) = cos(3t), y(t) = sin(2t)')

# Настройка заголовка и подписей осей
plt.title('Параметрическая кривая', fontsize=16)
plt.xlabel('x(t) = cos(3t)', fontsize=14)
plt.ylabel('y(t) = sin(2t)', fontsize=14)

# Добавление сетки
plt.grid(color='#EFB509', alpha=0.5)

# Добавление легенды
plt.legend()

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
years = np.array([2012, 2013, 2014, 2015, 2016,
                  2017, 2018, 2019, 2020, 2021, 2022])
minimum_income = np.array([4341.15, 4599.21, 4981.51, 5182.99, 4794.34,
                           4900.79, 4792.54, 4806.32, 5358.53,
                           6268.31, 7178.32])



# Построение столбиковой диаграммы
plt.figure(figsize=(12, 6))
bars = plt.bar(years, minimum_income, color='#EFB509',
               alpha=0.8, edgecolor='#002C54', linewidth=1.2)

# Настройка заголовка и подписей осей
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)',
          fontsize=18, fontweight='bold', color='#002C54')
plt.xlabel('Годы', fontsize=14, color='#002C54')
plt.ylabel('Прожиточный минимум (в сомах)', fontsize=14, color='#002C54')

# Добавление значений над столбиками
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50,
             f'{yval:.2f}', ha='center', va='bottom', fontsize=12, fontweight='bold', color='#002C54')

# Настройка внешнего вида графика
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(years, fontsize=12, color='#002C54')  # Устанавливаем метки по оси X
plt.yticks(fontsize=12, color='#002C54')

# Установим лимиты по оси Y для лучшего отображения значений
plt.ylim(4000, 7500)

# Отображение графика
plt.tight_layout()
plt.show()
```

## Задание 9

Построить график произвольной каркасной поверхности. 

### Описание

- **Каркасная поверхность**: График, состоящий из линий, соединяющих заданные точки в трехмерном пространстве.
- **Используемые библиотеки**: Для визуализации использовать `matplotlib` и `numpy`.
- **Пример**: Можно использовать функции для создания интересной каркасной поверхности, например, функцию синуса и косинуса.


```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
years = np.array([2012, 2013, 2014, 2015, 2016,
                  2017, 2018, 2019, 2020, 2021, 2022])
minimum_income = np.array([4341.15, 4599.21, 4981.51, 5182.99, 4794.34,
                           4900.79, 4792.54, 4806.32, 5358.53,
                           6268.31, 7178.32])

# Нахождение индекса максимального значения
max_index = np.argmax(minimum_income)

# Создание массива для explode
explode = [0.2 if i == max_index else 0 for i in range(len(minimum_income))]

# Построение кругового графика
plt.figure(figsize=(10, 10))
wedges, texts, autotexts = plt.pie(minimum_income, autopct='%1.1f%%',
                                   startangle=140, colors=plt.cm.Paired.colors, explode=explode)

# Настройка заголовка
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)',
          fontsize=18, fontweight='bold')

# Настройка текста на графике
for text in texts:
    text.set_fontsize(12)
    text.set_color('black')
for autotext in autotexts:
    autotext.set_color('#002C54')
    autotext.set_fontsize(12)

# Добавление легенды
plt.legend(bbox_to_anchor=(-0.16, 0.65, 0.25, 0.25),
           loc='lower left', labels=years)

plt.axis('equal')  # Равные оси, чтобы круг был правильным
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация сетки параметров
u = np.linspace(0, 2 * np.pi, 30)  # угол
v = np.linspace(-1, 1, 10)          # уменьшенный диапазон высоты
U, V = np.meshgrid(u, v)            # Создание двумерных сеток

# Определение координат X, Y, Z для гиперболоидной поверхности
x = np.cosh(V) * np.cos(U)
y = np.cosh(V) * np.sin(U)
z = V

# Создание графика
fig = plt.figure(figsize=(10, 8))  # Увеличение размера графика
ax = fig.add_subplot(111, projection='3d')

# Построение каркасной поверхности
ax.plot_wireframe(x, y, z, color='#002C54')

# Настройка осей
ax.set_xlabel('X', color='red')
ax.set_ylabel('Y',color='red')
ax.set_zlabel('Z',color='red')

# Настройка пределов осей для лучшего визуального восприятия
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-1, 1])  # Соответствующий диапазон для Z

# Отображение графика
plt.title('Wireframe of Hyperboloid Surface', color='#002C54')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация сетки параметров с использованием mgrid
# u - угол, v - уменьшенный диапазон высоты
u, v = np.mgrid[0:2 * np.pi:20j, -1:1:10j]

# Определение координат X, Y, Z для гиперболоидной поверхности
x = np.cosh(v) * np.cos(u)
y = np.cosh(v) * np.sin(u)
z = v

# Создание графика
fig = plt.figure(figsize=(10, 8))  # Увеличение размера графика
ax = fig.add_subplot(111, projection='3d')

# Построение поверхности
ax.plot_surface(x, y, z, cmap='inferno')

# Настройка осей
ax.set_xlabel('X', color='red')
ax.set_ylabel('Y', color='red')
ax.set_zlabel('Z', color='red')

# Настройка пределов осей для лучшего визуального восприятия
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-1, 1])  # Соответствующий диапазон для Z

# Отображение графика
plt.title('Surface of Hyperboloid', color='#002C54')
plt.show()
```
