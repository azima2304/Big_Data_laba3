## Задание 2

Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей и заголовка фигуры.

   Была взята функция вида:

   $$
   y = x^2 - 2x - 8
   $$

### Описание

- **Квадратичная функция**: Это функция вида $y = x^2 - 2x - 8$, где \( a, b, c \) — коэффициенты.
- **Масштаб**: Выбор масштаба осей может быть произвольным, чтобы лучше отразить характер графика.
- **Подписи осей**: Необходимо указать метки для осей X и Y для лучшего понимания графика.
- **Заголовок фигуры**: Заголовок должен быть информативным и отражать суть графика.


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Comic Sans MS'
mpl.rcParams['font.size'] = 12
mpl.rcParams['font.weight'] = 'bold'


fig, ax = plt.subplots(figsize=(10, 6))
ax.grid(color='#EFB509', alpha=0.5)

x = np.linspace(-5, 7, 100)
y = np.power(x, 2) - 2 * x - 8

ax.plot(x, y, color='#002C54', linestyle='-',
        linewidth=2, label='y = x² - 2x - 8')

plt.title('Quadratic Function: y = x² - 2x - 8', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)


ax.grid(True)
ax.legend()
plt.savefig('quadratic_function.png', format='png')
plt.show()
```

## Задание 3

Построить график произвольной гармонической функции, такой как

$$
y = A \sin(Bx + C) + D,
$$

с произвольным масштабом и настроенными подписями осей и заголовком. 

В этом примере мы возьмём гармоническую функцию 

$$
y = 2 \sin\left(2x + \frac{\pi}{4}\right) + 1.
$$

### Описание

- **Гармоническая функция**: Это функция, представляющая собой синусоиду с амплитудой \( A \), периодом $ \frac{2\pi}{B} $, фазовым сдвигом \( C \) и вертикальным сдвигом \( D \).
- **Масштаб**: Масштаб графика может быть произвольным для лучшего визуального восприятия.
- **Подписи осей**: Важно настроить метки для осей X и Y, чтобы отразить информацию о переменных.
- **Заголовок фигуры**: Заголовок должен быть ясным и информативным, описывающим суть графика.


```python
import matplotlib.pyplot as plt
import numpy as np

# Определяем диапазон x
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)  # 1000 точек от -2π до 2π

# Определяем гармоническую функцию
A = 2  # Амплитуда
B = 2  # Частота
C = np.pi / 4  # Фаза
D = 1  # Смещение по оси Y
y = A * np.sin(B * x + C) + D  # Гармоническая функция

# Создание фигуры
plt.figure(figsize=(12, 6))

# Построение графика
plt.plot(x, y, color='blue', linestyle='-',
         linewidth=2, label='y = 2sin(2x + π/4) + 1')

# Настройка заголовка и подписей осей
plt.title('График гармонической функции', fontsize=16)
plt.xlabel('Ось X (радианы)', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)

# Установка произвольного масштаба для осей
plt.xlim(-7, 7)  # Масштаб по оси X
plt.ylim(-3, 5)  # Масштаб по оси Y

# Добавление сетки
plt.grid(color='#EFB509', alpha=0.5)

# Добавление легенды
plt.legend()

# Отображение графика
plt.show()
```
